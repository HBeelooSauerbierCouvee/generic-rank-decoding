```python
decim_prec = 50
ω = 2

def red_zero(a, param_cost_list):  #a in range(k)
    q,m,n,k,r = param_cost_list[0]
    return([q, m, n-a, k-a, r], N(a*r*log(q,2) ,decim_prec) + (param_cost_list[1]) ) 


def red_row_supp(a_prime, param_cost_list): #a_prime in range(k)
    q,m,n,k,r = param_cost_list[0]
    extra = 0
    if a_prime > 0:
        extra = 2
    return([q, m, n, k-a_prime, r], N(log(max(n*a_prime,1),2) ,decim_prec) + N((a_prime*r+extra)*log(q,2) ,decim_prec)  + (param_cost_list[1]) ) 


def red_puncture(p, param_cost_list):  #p in range(max(n-k-r,1))
    q,m,n,k,r = param_cost_list[0]
    return([q, m, n-p, k, r], param_cost_list[1]) 


def AGHT(param_cost_list):
    q,m,n,k,r = param_cost_list[0]
    return(N(log( max(  ((n-k)^ω)*(m^ω)*q^(max(r*ceil((k+1)*m/n)-m, 0))  ,0.00001), 2),decim_prec) + param_cost_list[1])


def OJ2(param_cost_list):
    q,m,n,k,r = param_cost_list[0]
    return(N(log( max( ((r-1)*(m-1)+k+(r-1)+ceil((k+1)/(m-1))+1)^ω*q^((r-1)*(k+ceil((k+1)/(m-1))) + 2) ,0.00001), 2),decim_prec) + param_cost_list[1])


def GRS4(param_cost_list):
    q,m,n,k,r = param_cost_list[0]
    return(N(log(  max((r^ω)*(k^ω)*q^(max(r*ceil(((r+1)*(k+1)-(n+1))/r),0)),0.00001), 2),decim_prec) + param_cost_list[1])


def MM(param_cost_list):
    q,m,n,k,r = param_cost_list[0]
    current_cost = Infinity
    
    if m*binomial(n-k-1,r) >= binomial(n,r)-1:
        current_cost = N(log( max( m*binomial(n-k-1,r)*binomial(n,r)^(ω-1),0.00001), 2),decim_prec)
   
    return([q,m,n,k,r],current_cost + param_cost_list[1])


def SM(b, param_cost_list): #b in range(1,n-k-1+1)
    q,m,n,k,r = param_cost_list[0]
    current_cost = Infinity
    
    NFqm = N(-binomial(n-k-1,r)*binomial(k+b-1,b) + sum([binomial(n-i,r)*binomial(k+b-1-i,b-1) for i in range(1,k+1)]))
    NFqsyz = N((m-1)*sum([(-1)^(i+1)*binomial(n-k-1,r+i)*binomial(k+b-i-1,b-i) for i in range(1,b+1)]))
    NFq = N(NFqm - NFqsyz)
    MFq = N(binomial(k+b-1,b)*(binomial(n,r) - m*binomial(n-k-1,r)))
    if NFq >= MFq -1 and MFq >=1:
        current_cost = N(log( max(m^2*NFq*MFq^(ω-1),0.00001), 2),decim_prec)
    
    return([q,m,n,k,r], current_cost + param_cost_list[1])


def MM_test(params):
    q,m,n,k,r = params
    best_cost = Infinity
    for a_prime in range(k):
        for a in range(k - a_prime):
            for p in range(4):
                x = 2
                current_cost = MM(red_puncture(p, red_row_supp(a_prime, red_zero(a, ([q,m,n,k,r],0) ))))[1] 
                if current_cost <= min(best_cost, current_cost) and current_cost != Infinity:
                    best_cost = current_cost
                    print(f"a'={a_prime}, a={a}, p={p}", ceil(current_cost))
    print('best:', best_cost,'\n')
    
print(f"ω = {ω}")
```
